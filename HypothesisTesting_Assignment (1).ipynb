Question 1

Null hypothesis:There is no significant difference in the diameter of cutlets between two units

Alternate Hypothesis: There is significant diiference between the diameter of cutlets between two units

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv('Cutlets.csv')
df.head()
```

```python
unitA=pd.Series(df['Unit A'])
unitB=pd.Series(df['Unit B'])
```

```python
stats.ttest_ind(unitA,unitB)
```

Sinc p-value is greater than 0.05 we accepts the Null Hypothesis

i.e:There is no signifiacnt difference between the diameter of 2 cutlets

Question 2

Null Hypothesis:There is no significant difference in average TAT of reports of the laboratories

Alternate Hypothesis:There is significant difference in the average TAT of reports of the laboratories

```python
from google.colab import files
uploaded=files.upload()
```

```python
lab=pd.read_csv('LabTAT.csv')
lab.head()
```

```python
stats.f_oneway(lab['Laboratory 1'],lab['Laboratory 2'],lab['Laboratory 3'],lab['Laboratory 4'])
```

Since p-value<0.05,We can conclude that the Alternate Hypothesis is true

i.e:There is significant difference between the TAT of reports of laboratory
